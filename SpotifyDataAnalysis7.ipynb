# **Data Analysis**

1. Top 10 Popular Artists

```python
#Top Artist
Top_Artists = df.groupby("Artist").count().sort_values(by="Popularity", ascending=False) ["Popularity"][:10]
Top_Artists
```

```python
#Visualizing 
Top_Artists.plot.barh()
plt.show()
```

2. Top 10 Genres released in each year

```python
# Top 10 Genres released in each year 
Genre_yr = df.groupby("Genre").count().sort_values(by="Year", ascending=False) ["Year"][:10]
Genre_yr
```

```python
#Visualising
Genre_yr.plot.barh()
plt.show()
```

3.Top 5 Loudest Track

```python
#Top 5 Loudest Track
Loudest_Track= df[["Loudness","Title"]].sort_values(by="Loudness", ascending=True)[:5]
Loudest_Track
```

```python
#Visualising
sns.barplot(x="Loudness", y='Title', data=Loudest_Track, estimator=np.mean)
```

4.Top 5 Acousticness Tracks



```python
#Top 5 Acousticness Tracks
Top_Acousticness= df[["Acousticness","Title"]].sort_values(by="Acousticness", ascending=False)[:5]
Top_Acousticness
```

```python
#Visualising
sns.barplot(x="Acousticness", y='Title', data=Top_Acousticness, estimator=np.mean)
```

5.Danceability song

```python
#Danceability song
Danceability_Tracks= df[["Danceablity","Title"]].sort_values(by="Danceablity", ascending=False)[:5]
Danceability_Tracks
```

```python
#Visualising
sns.barplot(x="Danceablity", y='Title', data=Danceability_Tracks, estimator=np.mean)
```

6. Top 5 Energetic tracks

```python
#Top 5 Energetic tracks
Energetic_Tracks= df[["Energy","Title"]].sort_values(by="Energy", ascending=False)[:5]
Energetic_Tracks
```

```python
#Visualising
plt.figure(figsize=(12,7))
plt.pie(x="Energy", data=Energetic_Tracks, autopct='%1.2f%%',labels=Energetic_Tracks.Title)
plt.show()
```

7. Number of Artist released their songs in each year

```python
# Number of Artist released their songs in each year
Artist_yr = df.groupby("Year").count().sort_values(by="Artist", ascending=False) ["Artist"]
Artist_yr
```

```python
#Visualising 
Artist_yr.plot.bar()
plt.show()
```

8.Top 10 songs with good tempo (120 Tempo is Perfect)

```python
#Top 10 songs with good Tempo
tempo_filtered = df[(df["Tempo"] >= 100) & (df["Tempo"] <= 140)]
top_10_tempos = tempo_filtered.groupby("Title")["Tempo"].apply(list).reset_index(name="Tempo")[:10]
top_10_tempos
```

```python
#Visualising 

# Extracting the title and tempo columns from the DataFrame
titles = top_10_tempos["Title"]
tempos = top_10_tempos["Tempo"].apply(lambda x: x[0])

# Creating a bar plot
plt.figure(figsize=(10, 6))
plt.bar(titles, tempos)
plt.xlabel("Title")
plt.ylabel("Tempo")
plt.title("Top 10 Titles with Tempos between 100 and 140")
plt.xticks(rotation=90)  # Rotate x-axis labels for better visibility
plt.show()


```
