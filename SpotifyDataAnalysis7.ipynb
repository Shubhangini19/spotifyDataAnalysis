# **SUMMARY**




The Spotify analysis dataset contains a collection of data related to music tracks available on the Spotify platform. The dataset includes various attributes and features of the songs, such as their popularity,Livness , tempo, energy, danceability, Valence, and more. This data can be used to gain insights into music trends, preferences, and characteristics.

Here is a summary of the key information you can find in the Spotify analysis dataset:


1.   Track Information: The dataset provides details about individual tracks, including the track name, artist name, album name, release year.
2.   Audio Features: Spotify calculates a set of audio features for each track, which are included in the dataset. These features include danceability (how suitable a track is for dancing), energy (perceived intensity and activity), loudness, tempo and more. These features help describe the musical characteristics of the songs.


3.   Popularity: The dataset includes a popularity score for each track, indicating how popular it is among Spotify users. The popularity score ranges from 0 to 100, with higher values representing greater popularity.
4.   Genre: Some datasets may include information about the genre of the track. This label can provide insights into the style or emotional qualities of the music.

By analyzing this dataset, researchers and music enthusiasts can explore various aspects of the music industry, such as identifying popular genres or trends, understanding the impact of different musical features on track popularity, and creating personalized music recommendations based on user preferences.

8.Top 10 songs with good tempo (120 Tempo is Perfect)

```python
#Top 10 songs with good Tempo
tempo_filtered = df[(df["Tempo"] >= 100) & (df["Tempo"] <= 140)]
top_10_tempos = tempo_filtered.groupby("Title")["Tempo"].apply(list).reset_index(name="Tempo")[:10]
top_10_tempos
```

```python
#Visualising

# Extracting the title and tempo columns from the DataFrame
titles = top_10_tempos["Title"]
tempos = top_10_tempos["Tempo"].apply(lambda x: x[0])

# Creating a bar plot
plt.figure(figsize=(10, 6))
plt.bar(titles, tempos)
plt.xlabel("Title")
plt.ylabel("Tempo")
plt.title("Top 10 Titles with Tempos between 100 and 140")
plt.xticks(rotation=90)  # Rotate x-axis labels for better visibility
plt.show()


```

### **Observation:**


*   7 Years has the most perfect Tempo which is 120.
*   A Thousand Years also have good Tempo.



9.Correlation between the variables

```python
#Correlation between the variables
# Convert non-numeric values to NaN
df = df.apply(pd.to_numeric, errors='coerce')

# Calculate correlation matrix
correlation_matrix = df.corr()

# Create heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

### **Observation:**


*  Energy is high while Acousticness is low they are least correlated.
*  Valence and Danceability is more correlated.



10.Negative and Positive Values in the dataset

```python
#Negative and Positive Values in the dataset
# Plot separate histograms for each column
num_cols = df.shape[1]  # Number of columns
fig, axes = plt.subplots(nrows=num_cols, ncols=1, figsize=(8, 6*num_cols))

for i, col in enumerate(df.columns):
    positive_data = df[col][df[col] >= 0]
    negative_data = df[col][df[col] < 0]
    axes[i].hist(positive_data, bins=10, color='blue', alpha=0.5, label='Positive')
    axes[i].hist(negative_data, bins=10, color='red', alpha=0.5, label='Negative')
    axes[i].set_xlabel('Value')
    axes[i].set_ylabel('Frequency')
    axes[i].set_title(f'Histogram - {col}')
    axes[i].legend()

plt.tight_layout()
plt.show()
```

### **Observation:**


*   Loudness is the column with negative Values.
*   Other columns have postive Values.
